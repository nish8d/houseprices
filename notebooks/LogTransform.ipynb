```python
import joblib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load preprocessed data
X_train = joblib.load('../data/X_train.pkl')
X_test = joblib.load('../data/X_test.pkl')
y_train = joblib.load('../data/y_train.pkl')
y_test = joblib.load('../data/y_test.pkl')
```

```python
# Plot original target distribution
sns.histplot(y_train, kde=True)
plt.title("Original SalePrice Distribution")
plt.show()
```

```python
# Apply log1p transformation
y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)
```

```python
# Plot transformed target distribution
sns.histplot(y_train_log, kde=True)
plt.title("Log-Transformed SalePrice Distribution")
plt.show()
```

```python
joblib.dump(y_train_log, 'D:/datascienceprojects/houseprices/data/y_train_log.pkl')
joblib.dump(y_test_log, 'D:/datascienceprojects/houseprices/data/y_test_log.pkl')
```
